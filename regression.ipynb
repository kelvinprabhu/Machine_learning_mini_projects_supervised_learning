```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
hs = np.array([1500, 1700, 2100, 2400, 3000])
p = np.array([300000, 320000, 340000, 360000, 400000])
model = LinearRegression()
model.fit(hs.reshape(-1, 1), p.reshape(-1, 1))
```

```python
plt.scatter(hs, p)
plt.plot(hs, model.predict(hs.reshape(-1, 1)), color='red')
```

```python
slope = model.coef_[0][0]
intercept = model.intercept_[0]
print(slope)
print(intercept)
pred = slope * 2500 + intercept
print('price of a house with 2500 sq ft: ',pred)
```

```python
plt.scatter(hs, p)
plt.plot(hs, model.predict(hs.reshape(-1, 1)), color='red')
plt.scatter(2500, pred, color='green')
```
